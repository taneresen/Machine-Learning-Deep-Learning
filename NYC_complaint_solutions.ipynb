```python
# first read data from the csv file into a dataframe
import pandas as pd
filepath="/Users/sefer_imac/Google Drive/dersler/python-problems/python_pandas/NYC_complaint/311-service-requests.csv"
data=pd.read_csv(filepath,index_col=0)
data.head()
```


### Find average completion time in hours for each complaint types for top-10 most frequent complaints

```python
# first change the data types of Created Date and Closed data columns to datetime type 
data["Created Date"]=pd.to_datetime(data["Created Date"])
data["Closed Date"]=pd.to_datetime(data["Closed Date"])
```

```python
# extract rows for top10 most frequent complaint types
top10_complaint=data["Complaint Type"].value_counts()[:10]
data_top10_complaint=data[data["Complaint Type"].isin(top10_complaint.index)]
```

```python
# Extract only closed complaints
data_top10_complaint_closed=data_top10_complaint[data_top10_complaint["Status"]=="Closed"].copy()
```

```python
# calculate closing time: difference between closed date and created datae
data_top10_complaint_closed["closing_time"]=data_top10_complaint_closed["Closed Date"]-data_top10_complaint_closed["Created Date"]
```

```python
# create closing_time_hours column showing total hours spent for closing
data_top10_complaint_closed["closing_time_hours"]=data_top10_complaint_closed.closing_time.map(lambda x: (x.total_seconds())/3600.0)
```

```python
# now calculate mean closing time and number of complaints for each complaint type
data_top10_complaint_closed.pivot_table("closing_time_hours",index="Complaint Type",aggfunc=["mean","count"]).round(2)
```

### produce a table for the number complaint type in each day of the week.
Do this analysis only for the top-10 most frequent complaints. 

First, find the top-10 most frequent complaint type. Extract data just for these 10-most frequent complaint types. Then, produce a table for the number complaint type in each day of the week.

```python
# first the read the file
import pandas as pd
data=pd.read_csv(filepath)
```

```python
# change the data types of Created Date and Closed data columns to datetime type 
data["Created Date"]=pd.to_datetime(data["Created Date"])
data["Closed Date"]=pd.to_datetime(data["Closed Date"])
```

```python
# create day column to show the day name of the datetime for example monday,sunday etc
data["day"]=data["Created Date"].map(lambda x: x.strftime("%A"))
```

```python
# extract rows for the top-10 most frequent complaint types
top10_complaint=data["Complaint Type"].value_counts()[:10]
data_top10_complaint=data[data["Complaint Type"].isin(top10_complaint.index)]
```

```python
# now calculate the number of complaints for each day and for each complaint type
# show result for complaint types on columns
data_top10_complaint.pivot_table(index="day",columns="Complaint Type",aggfunc="size")
```

### find the percentages for the status of Complaints.

Use data for top-10 most frequent complaint types

```python
# first the read the file
import pandas as pd
data=pd.read_csv(filepath)
```

```python
# extract rows for the top-10 most frequent complaint types
top10_complaint=data["Complaint Type"].value_counts()[:10]
data_top10_complaint=data[data["Complaint Type"].isin(top10_complaint.index)]
```

```python
# write calc_percentage function which takes a data frame and returns percentages for Status column
def calc_percentage(x):
    return x.Status.value_counts(normalize=True)*100
```

```python
# now apply this function for each complaint type 
# show each status as a seperate column
# round the data for two decimals
# replace NAN values with 0
data_top10_complaint.groupby("Complaint Type").apply(calc_percentage).unstack(level=1).fillna(0).round(2)
```
