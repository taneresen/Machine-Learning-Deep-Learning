```python
df=pd.read_csv('weight-height.csv')
```

```python
df
```

```python
fig = sns.scatterplot(data=df, x='Weight',y='Height',hue='Gender')
```

```python
#create a dataframe with all training data except the target column
X = df.drop(columns=['Gender'])
#check that the target variable has been removed
X.head()
```

```python
#separate target values
y = df['Gender'].values
#view target values
y[0:5]
```

```python
from sklearn.neighbors import KNeighborsClassifier
# Create KNN classifier
# KNeighborsClassifier() yaparak icinde shift - tab yapinca neler yapilir gÃ¶rÃ¼lÃ¼r
knn = KNeighborsClassifier(n_neighbors = 3)
# Fit the classifier to the data
knn.fit(X,y)
```

```python
#Warning: This is not a correct way of training a model, the data should be split into training and test set
#This is cheating
knn.score(X, y)
```

```python
#split dataset into train and test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1, stratify=y)
```

```python
# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors = 3)
# Fit the classifier to the data
knn.fit(X_train,y_train)
```

```python
#Make predictions on the test data
predictions=knn.predict(X_test)
```

```python
list(zip(predictions,y_test))
```

```python
#check accuracy of our model on the test data
knn.score(X_test, y_test)
```
