# Scatter Plots

A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables (Cartio.com).

In this part of our class, we will use surface area and Population Data for  Countries from Worldbank. As you can guess, Surface Area and Population will be our two variables.  

-Iki degiskenin bir biri arasindaki iliskiyi göstermek icin kullanilir.

````python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Read world bank Area Data -Yüz ölcümlerinin oldugu dosya
df_Area = pd.read_excel("data_files\Worldbank Country Area.xlsx",index_col=0)

df_Area.tail()
````

````python
#Read world bank Population Data-nufusun oldugu veri
df_Pop = pd.read_excel("data_files\Worldbank Population Data.xlsx",index_col=0)

df_Pop.tail()
````

````python
#List of Some Euroasian countries 
Countries = ["Belgium", "Bulgaria", "Germany","Denmark","Spain","Finland","France","United Kingdom",
             "Italy","Netherlands","Norway", "Poland", "Portugal","Sweden","Turkey","Ukraine"]

#let's get 2017 surface area and population data for these countries
x_values = df_Area.loc[2017,Countries]/1000 # express in thousands of km2
y_values =  df_Pop.loc[2017,Countries]/1000000 #express in millions

#create scatter plot
plt.figure(dpi=150)
plt.scatter(x_values,y_values)

plt.title("Population - Surface Area")
plt.xlabel("Surface Area (x 1000 km2)")
plt.ylabel("Population (in millions)")
plt.show()
````

### Figure Annotation

As we can see in the above plots, countries to which each data point belong to is not shown. Which makes plots harder to read. We may want to add a small explanataion next to some or all data points in similar situations. We can achieve this by using ```pyplot.annotate()``` 

````python
#create scatter plot
plt.figure(dpi=150)

plt.scatter(x_values,y_values)

plt.title("Population - Surface Area")
plt.xlabel("Surface Area (x 1000 km2)")
plt.ylabel("Population (in millions)")


for country in Countries:
    plt.annotate(country, (x_values[country],y_values[country]),fontsize=6) #try: fontsize=5
    
    
plt.show()
````

### Annotation offset and arrow styles

**Arrow styles and other parameters:** https://matplotlib.org/3.2.2/api/text_api.html#matplotlib.text.Text 

````python
#create scatter plot
plt.figure(dpi=150)

plt.scatter(x_values,y_values)

plt.title("Population - Surface Area")
plt.xlabel("Surface Area (x 1000 km2)")
plt.ylabel("Population (in millions)")


for country in Countries:
    plt.annotate(country, (x_values[country],y_values[country]),
                 xytext=(15,10),
                 textcoords = 'offset pixels',
                 fontsize= 5, arrowprops = {'arrowstyle': '->'})
    
    
plt.show()
````

### Use different colors within the same data series

if we supply parameter *c* (an array of integers) to scatter plot. Each integer will be mapped to a color in the current colormap. Then each data point will have the next color in line.

````python
#create random numbers for random colors 
colors = np.random.randint(0,256,len(Countries))

colors2 =[ "#FF5733","#33FF3D","#FD33FF","#FF3345","#581845",
           "#FF5733","#33FF3D","#FD33FF",
           "#FF3345","#581845","#FF5733",
           "#33FF3D","#FD33FF","#FF3345","#581845"]

#create scatter plot
plt.figure(dpi=150)

plt.scatter(x_values,y_values,c = colors)

plt.title("Population - Surface Area")
plt.xlabel("Surface Area (x 1000 km2)")
plt.ylabel("Population (in millions)")


for country in Countries:
    plt.annotate(country, (x_values[country],y_values[country]),
                 xytext=(15,10),
                 textcoords = 'offset pixels',
                 fontsize= 5, arrowprops = {'arrowstyle': '->'})
    
    
plt.show()
````

### Use different colormaps 

This applies to all kinds of figures we can create with Matplotlib. We can change to colormap of the figure by changing *cmap* parameter.

**full list of colormaps:** https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html

````python
#create random numbers for random colors 
np.random.seed(10)
colors = np.random.randint(0,256,len(Countries))

#create scatter plot
plt.figure(dpi=150)

plt.scatter(x_values,y_values,c = colors,cmap= 'Spectral')
#plt.scatter(x_values,y_values,c = colors,cmap= 'twilight')
#plt.scatter(x_values,y_values,c = colors,cmap= 'Greens')


plt.title("Population - Surface Area")
plt.xlabel("Surface Area (x 1000 km2)")
plt.ylabel("Population (in millions)")


for country in Countries:
    plt.annotate(country, (x_values[country],y_values[country]),
                 xytext=(15,10),
                 textcoords = 'offset pixels',
                 fontsize= 5, arrowprops = {'arrowstyle': '->'})
    
    
plt.show()
````

### Adding a third dimension to a Scatter Plot (i.e. Create a Bubble Plot) 

We can create a bubble plot by supplying individual marker sizes to each data point. We can achieve this by setting parameter *s* for scatter plot equal to an array containing the size of individual data points. 

Let's make the size of each data point proportional to the corresponding country's GDP. Hence the third dimension in our figure will be the GDP of the country. 

````python
#Read world bank GDP Data
df_GDP = pd.read_excel("data_files\Worldbank GDP Data.xlsx",index_col=0)

#Fill missing values with 0 
df_GDP = df_GDP.fillna(0)

#express in billions of dollars
df_GDP = df_GDP /10000000000

df_GDP.tail()
````

````python
#get 2017 GDP for relevant countries

gdps= df_GDP.loc[2017,Countries]

gdps
````

````python
#create random numbers for random colors 
np.random.seed(10)
colors = np.random.randint(0,100,len(Countries))

#create scatter plot
plt.figure(dpi=150)

#add size information here
#add alpha to make circles transparent
plt.scatter(x_values,y_values,c = colors, s=gdps, cmap= 'Spectral', alpha = 0.5)



plt.title("Population - Surface Area") 
plt.xlabel("Surface Area (x 1000 $km^2$)") #$km^{20}$
plt.ylabel("Population (in millions)")


for country in Countries:
    plt.annotate(country, (x_values[country],y_values[country]),
                 xytext=(15,10),
                 textcoords = 'offset pixels',
                 fontsize= 5, arrowprops = {'arrowstyle': '->'})
    

#change limits to make sure all bubbles fit in the graph
plt.xlim(-50,900)
plt.ylim(-5,100)

#plt.yticks([0,5,10,15,20,40,60,80,85,90,95])
    
plt.show()
````
