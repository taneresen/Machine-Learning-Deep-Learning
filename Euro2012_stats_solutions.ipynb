### Analysis on Euro 2012 statistics 

```python
# first read 'Euro 2012 stats TEAM.csv' file into a dataframe
import pandas as pd
data=pd.read_csv('Euro 2012 stats TEAM.csv')
data.head()
```

#### Find the teams that have lowest and highest number of yellow cards. 
Note that more than one country can have the same maximum number or minimum number of yellow cards

```python
# first find the number of minimum yellow card
min_yellow=data["Yellow Cards"].min()
```

```python
# now find rows where we see Yellow card number is the minimum number
countries_min_yellow=data[data["Yellow Cards"]==min_yellow]
countries_min_yellow.head()
```

```python
# now print the countries
print("countries having the mininum number of yellow cards:\n",*countries_min_yellow.Team)
```

- we can to the same for the maximum number of yellow cards

```python
max_yellow_card=data[data["Yellow Cards"]==data["Yellow Cards"].max()].Team
print("countries having the maximum number of yellow cards:\n",*max_yellow_card)
```

### Find the team that has the best shooting accuracy

```python
# Shooting Accuracy column has the percentage character
# First delete the percentage character and change the Shooting Accuracy column into floating points
data["Shooting Accuracy"]=data["Shooting Accuracy"].str.replace("%","").astype(float)
```

```python
# now follow a different strategy
# set the Team column as the index
data.set_index("Team",inplace=True)
data.head()
```

```python
# let's check if the index is unique or not
data.index.is_unique
```

```python
# now find the index of the highest Shooting Accuracy
data["Shooting Accuracy"].idxmax()
```
