# Matplotlib Plot Style Sheets

### Matplotlib rcParams
Matplotlib uses matplotlibrc configuration files to customize all kinds of properties, which we call 'rc settings' or 'rc parameters'. You can control the defaults of almost every property in Matplotlib: figure size and DPI, line width, color and style, axes, axis and grid properties, text and font properties and so on. Matplotlib looks for matplotlibrc in four locations, in the following order:

1. matplotlibrc in the current working directory, usually used for specific customizations that you do not want to apply elsewhere.

2. "[dollar_sign]MATPLOTLIBRC" if it is a file, else [dollar_sign]MATPLOTLIBRC/matplotlibrc.

3. It next looks in a user-specific place, depending on your platform: On Linux and FreeBSD, it looks in .config/matplotlib/matplotlibrc (or [dollar_sign]XDG_CONFIG_HOME/matplotlib/matplotlibrc) if you've customized your environment. On other platforms, it looks in .matplotlib/matplotlibrc. See matplotlib configuration and cache directory locations.

4. INSTALL/matplotlib/mpl-data/matplotlibrc, where INSTALL is something like /usr/lib/python3.7/site-packages on Linux, and maybe C:\Python37\Lib\site-packages on Windows. Every time you install matplotlib, this file will be overwritten, so if you want your customizations to be saved, please move this file to your user-specific matplotlib directory.

### Changing the style sheet

Use ```pyplot.style.use(sytle_name)``` function to switch to an available style. If you have more than one user defined style sheets, you can combine them as ```pyplot.style.use([user_style1,user_style2,...])```

Let's see one of the plots we made earlier with two different styles:

````python
import matplotlib.ticker as ticker
empCount = [82, 91, 105, 112, 111, 109, 110, 110]
revenue= [0.6, 0.5, 0.7, 0.9, 0.6, 0.6, 0.8, 1]

fig, ax1 = plt.subplots()
ax1.bar(np.arange(8), revenue)
ax1.set_xticks(np.arange(8))
ax1.set_xticklabels([ "13Q1", "13Q2", "13Q3", "13Q4", "14Q1", "14Q2", "14Q3", "14Q4"])
ax1.set_ylabel("Revenue (Millions)")
ax1.set_ylim(0,1.2)
#for adding $ sign to y-axis
formatter = ticker.FormatStrFormatter('$%1.1f')
ax1.yaxis.set_major_formatter(formatter)


#copy axes 1 to create axes2
ax2 = ax1.twinx()

ax2.plot(empCount,marker='o')
ax2.set_ylabel("# of Sales Employees")
ax2.set_ylim(0,120)
ax2.set_label("Size of Salesforce")

fig.set_dpi(150)

plt.show()
````

````python
plt.style.use('fivethirtyeight')

fig, ax1 = plt.subplots()
ax1.bar(np.arange(8), revenue)
ax1.set_xticks(np.arange(8))
ax1.set_xticklabels([ "13Q1", "13Q2", "13Q3", "13Q4", "14Q1", "14Q2", "14Q3", "14Q4"])
ax1.set_ylabel("Revenue (Millions)")
ax1.set_ylim(0,1.2)
#for adding $ sign to y-axis
formatter = ticker.FormatStrFormatter('$%1.1f')
ax1.yaxis.set_major_formatter(formatter)


#copy axes 1 to create axes2
ax2 = ax1.twinx()

ax2.plot(empCount,marker='o')
ax2.set_ylabel("# of Sales Employees")
ax2.set_ylim(0,120)
ax2.set_label("Size of Salesforce")

fig.set_dpi(150)

plt.show()
````
